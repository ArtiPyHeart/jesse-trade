```python
import numpy as np

candles_1m = np.load("data/btc_1m.npy")
candles_1m.shape
```

```python
from custom_indicators.config import DOLLAR_BAR_THRESHOLD_10M
from custom_indicators.toolbox.dollar_bar import DollarBarContainer

dollar_bar_container_1 = DollarBarContainer(DOLLAR_BAR_THRESHOLD_10M, max_bars=500)
dollar_bar_container_2 = DollarBarContainer(DOLLAR_BAR_THRESHOLD_10M, max_bars=500)
```

```python
dollar_bar_container_1.load_initial_candles(candles_1m[:10000])
dollar_bar_container_2.load_initial_candles(candles_1m[:10000])
```

```python
candles_1m[-1]
```

```python
for i in range(10001, 50000):
    row = candles_1m[i]
    full = candles_1m[: i + 1]
    dollar_bar_container_1.update_with_candle(row)
    dollar_bar_container_2.update_with_candle(full)
    assert dollar_bar_container_1.bars.shape == dollar_bar_container_2.bars.shape
    assert (
        dollar_bar_container_1.is_latest_bar_complete()
        == dollar_bar_container_2.is_latest_bar_complete()
    )
    assert np.allclose(dollar_bar_container_1.bars, dollar_bar_container_2.bars)
```
