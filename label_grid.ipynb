```python
import plotly.express as px


def plot_distribution(res, bins=1000):
    fig = px.histogram(res, nbins=bins)
    fig.show()
```

```python
def get_side_label(
    candles,
    target_ret,
    pt_sl,
    last_ret,
    num_hours=0,
    num_minutes=0,
    z_score=2,
    verbose=False,
):
    labeler = TripleBarrierLabeler(
        candles,
        num_hours=num_hours,
        num_minutes=num_minutes,
        z_score=z_score,
        verbose=verbose,
    )
    side_labels = labeler.side_labels(pt=pt_sl, sl=pt_sl, target_ret=target_ret)
    candle_df = expand_labels(side_labels, candles, fill=0)
    ret = return_of_label(candle_df)
    label_counts = candle_df["bin"].value_counts().sort_index().to_numpy()
    print(
        f"{num_hours = } {num_minutes = } {target_ret = :.6f} {pt_sl = :.2f} -> {label_counts = } {ret = :.6f}({ret - last_ret :.6f})"
    )
    return ret
```

# meta model labels

```python
from jesse import helpers, research

from offline.labeling.labeling import (
    TripleBarrierLabeler,
    expand_labels,
    return_of_label,
)

_, trading_15m = research.get_candles(
    "Binance Perpetual Futures",
    "BTC-USDT",
    "15m",
    helpers.date_to_timestamp("2020-06-01"),
    helpers.date_to_timestamp("2025-01-31"),
    warmup_candles_num=0,
    caching=False,
    is_for_jesse=False,
)
```

```python
import lightgbm as lgb
import pandas as pd

from custom_indicators.selection import SIDE_ALL


def side_pred_to_label(pred):
    if pred > 0.6:
        return 1
    elif pred < 0.4:
        return -1
    else:
        return 0


df_features = pd.read_parquet("data/features_15m.parquet")

side_model = lgb.Booster(model_file="custom_indicators/models/model_side.txt")
side_model_pred = side_model.predict(df_features[SIDE_ALL])
df_features["model_side_res"] = side_model_pred


def get_meta_label(
    candles, target_ret, pt, sl, last_ret, num_minutes=0, num_hours=0, verbose=False
):
    labeler = TripleBarrierLabeler(
        candles, num_hours=num_hours, num_minutes=num_minutes, verbose=verbose
    )
    side_labels = labeler.side_labels(pt=pt, sl=sl, target_ret=target_ret)
    side_full_df = expand_labels(side_labels, candles, fill=0)
    side_full_df["pred"] = [side_pred_to_label(i) for i in side_model_pred]
    final_side_labels = side_full_df[side_full_df["ret"].notna()]
    meta_labels = labeler.meta_labels(
        final_side_labels, pt=pt, sl=sl, target_ret=target_ret
    )
    candle_df = expand_labels(meta_labels, candles, fill=0)
    ret = return_of_label(candle_df, meta_label=True)
    label_counts = candle_df["bin"].value_counts().sort_index().to_numpy()
    print(
        f"{num_hours = :.2f} {num_minutes = :.2f} {target_ret = :.6f} {pt = :.2f} {sl = :.2f} -> {label_counts = } {ret = :.6f}({ret - last_ret :.6f})"
    )
    return ret
```

```python
import numpy as np

LAST_RET = [0]
num_hours = 2.75
num_minutes = 0
target_ret = 0.0005
for ptsl in np.arange(1.0, 3.0, 0.1):
    pt = ptsl
    sl = ptsl
    ret = get_meta_label(
        trading_15m, target_ret, pt, sl, LAST_RET[-1], num_hours=num_hours
    )
    LAST_RET.append(ret)
```

```python
# final meta label
import numpy as np
import pandas as pd
import lightgbm as lgb
from jesse import helpers, research

from custom_indicators.selection import SIDE_ALL
from offline.labeling.labeling import TripleBarrierLabeler, expand_labels

_, trading_15m = research.get_candles(
    "Binance Perpetual Futures",
    "BTC-USDT",
    "15m",
    helpers.date_to_timestamp("2020-06-01"),
    helpers.date_to_timestamp("2025-01-31"),
    warmup_candles_num=0,
    caching=False,
    is_for_jesse=False,
)


def side_pred_to_label(pred):
    if pred > 0.6:
        return 1
    elif pred <= 0.4:
        return -1
    else:
        return 0


df_features = pd.read_parquet("data/features_15m.parquet")
side_model = lgb.Booster(model_file="custom_indicators/models/model_side.txt")
side_model_pred = side_model.predict(df_features[SIDE_ALL])
df_features["model_side_res"] = side_model_pred

num_hours = 1.8
num_minutes = 0
pt = 2.15
sl = 2.15
target_ret = 0.0008
labeler = TripleBarrierLabeler(
    trading_15m, num_hours=num_hours, num_minutes=num_minutes, verbose=True
)
side_labels = labeler.side_labels(pt=pt, sl=sl, target_ret=target_ret)
side_full_df = expand_labels(side_labels, trading_15m, fill=0, ffill=True)
side_full_df["pred"] = [side_pred_to_label(i) for i in side_model_pred]
final_side_labels = side_full_df[side_full_df["ret"].notna()]
meta_labels = labeler.meta_labels(
    final_side_labels, pt=pt, sl=sl, target_ret=target_ret
)
candle_df = expand_labels(meta_labels, trading_15m, fill=0, ffill=True)
print(candle_df.shape)
candle_df.head()
```
