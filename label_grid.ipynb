```python
from custom_indicators.toolbox.labeler.triple_barrier import (
    TripleBarrierLabeler,
    expand_labels,
    return_of_label,
)


def get_side_label(
    candles,
    target_ret,
    pt_sl,
    last_ret,
    num_hours=0,
    num_minutes=0,
    z_score=2,
    verbose=False,
):
    labeler = TripleBarrierLabeler(
        candles,
        num_hours=num_hours,
        num_minutes=num_minutes,
        z_score=z_score,
        verbose=verbose,
    )
    side_labels = labeler.side_labels(pt=pt_sl, sl=pt_sl, target_ret=target_ret)
    candle_df = expand_labels(side_labels, candles, fill=0)
    ret = return_of_label(candle_df)
    label_counts = candle_df["bin"].value_counts().sort_index().to_numpy()
    print(
        f"time={num_hours}:{num_minutes} {target_ret = :.6f} {pt_sl = :.2f} {z_score = :.2f} -> {label_counts = } {ret = :.6f}({ret - last_ret :.6f})"
    )
    return ret
```

# meta model labels

```python
import numpy as np
from jesse import helpers, research

from custom_indicators.toolbox.labeler.triple_barrier import (
    TripleBarrierLabeler,
    expand_labels,
    return_of_label,
)

trading_mid = np.load("data/label_dollar_bar_45m.npy")
trading_mid.shape
```

```python
import lightgbm as lgb
import pandas as pd

from custom_indicators.config import SIDE_ALL


def side_pred_to_label(pred):
    if pred > 0.5:
        return 1
    elif pred <= 0.5:
        return -1


df_features = pd.read_parquet("data/features_45m.parquet")

side_model = lgb.Booster(model_file="custom_indicators/models/model_side.txt")
side_model_pred = side_model.predict(df_features[SIDE_ALL])
df_features["model_side_res"] = side_model_pred


def get_meta_label(
    candles,
    target_ret,
    pt,
    sl,
    last_ret,
    num_minutes=0,
    num_hours=0,
    z_score=2,
    verbose=False,
):
    labeler = TripleBarrierLabeler(
        candles,
        num_hours=num_hours,
        num_minutes=num_minutes,
        z_score=z_score,
        verbose=verbose,
    )
    side_labels = labeler.side_labels(pt=pt, sl=sl, target_ret=target_ret)
    side_full_df = expand_labels(side_labels, candles, fill=0)
    side_full_df["pred"] = [side_pred_to_label(i) for i in side_model_pred]
    final_side_labels = side_full_df[side_full_df["ret"].notna()]
    meta_labels = labeler.meta_labels(
        final_side_labels, pt=pt, sl=sl, target_ret=target_ret
    )
    candle_df = expand_labels(meta_labels, candles, fill=0)
    ret = return_of_label(candle_df, meta_label=True)
    label_counts = candle_df["bin"].value_counts().sort_index().to_numpy()
    print(
        f"time={num_hours:.2f}:{num_minutes:.2f} {target_ret = :.6f} {pt = :.2f} {sl = :.2f} -> {label_counts = } {ret = :.6f}({ret - last_ret :.6f})"
    )
    return ret
```

```python
import numpy as np

LAST_RET = [0]
num_hours = 24
num_minutes = 0
target_ret = 0.001
for ptsl in np.arange(1.0, 3.1, 0.1):
    pt = ptsl
    sl = ptsl
    ret = get_meta_label(
        trading_mid[:, :6], target_ret, pt, sl, LAST_RET[-1], num_hours=num_hours
    )
    LAST_RET.append(ret)
```

```python
# final meta label
import lightgbm as lgb
import numpy as np
import pandas as pd
from jesse import helpers, research

from custom_indicators.config import SIDE_ALL
from custom_indicators.toolbox.labeler.triple_barrier import (
    TripleBarrierLabeler,
    expand_labels,
)
from custom_indicators.toolbox.labeler.triple_barrier_raw import TripleBarrierLabelerRaw

trading_mid = np.load("data/label_dollar_bar_45m.npy")


def side_pred_to_label(pred):
    if pred > 0.5:
        return 1
    elif pred <= 0.5:
        return -1
    else:
        return 0


df_features = pd.read_parquet("data/features_45m.parquet")
side_model = lgb.Booster(model_file="custom_indicators/models/model_side.txt")
side_model_pred = side_model.predict(df_features[SIDE_ALL])
df_features["model_side_res"] = side_model_pred

num_hours = 14
num_minutes = 0
pt = 2
sl = 2
target_ret = 0.002
z_score = 1.1
labeler = TripleBarrierLabelerRaw(
    trading_mid,
    num_hours=num_hours,
    num_minutes=num_minutes,
    z_score=z_score,
    verbose=True,
)
side_full_df = labeler.side_labels(pt=2.7, sl=2.7, target_ret=target_ret)
side_full_df = expand_labels(side_full_df, trading_mid, fill=0)
side_labels = [side_pred_to_label(i) for i in side_model_pred]
side_full_df["pred"] = [side_pred_to_label(i) for i in side_model_pred]
side_full_df
```
