```python
import pandas as pd
from sklearn.decomposition import PCA

df = pd.read_csv('RB99_1m_Turnover_31000_12120_1213.csv_tz80_Train_10877.csv')   ##### 训练集数据

features = df.drop('A0', axis=1)    ### 删除标签列

pca = PCA(n_components=50)  ### 降到指定维数

features_pca = pca.fit_transform(features)

row_index = 3000   ### 指定任意一行

print(features_pca[row_index])

single_row = features.iloc[[row_index]]  

single_row_pca = pca.transform(single_row)

### 查看单独降维后的结果
print('############################')
print(single_row_pca[0])

### 验证两种方式得到的结果是否一致
print(f"\n两种方式结果的差异总和: {sum(abs(features_pca[row_index] - single_row_pca[0]))}")
    
```
